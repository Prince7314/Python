<h2>1) Write a program to implement a calculator that performs basic operations (addition, subtraction, multiplication, division). 
Handle exceptions like division by zero, invalid operator input, etc. (A) </h2>

```python
class Calculator:
    def add(self, a, b):
        return a + b

    def subtract(self, a, b):
        return a - b

    def multiply(self, a, b):
        return a * b

    def divide(self, a, b):
        if b == 0:
            raise ZeroDivisionError("Cannot divide by zero.")
        return a / b


calc = Calculator()

try:
    a = float(input("Enter first number: "))
    b = float(input("Enter second number: "))
    op = input("Enter operator (+, -, *, /): ")

    if op == '+':
        print("Result:", calc.add(a, b))

    elif op == '-':
        print("Result:", calc.subtract(a, b))

    elif op == '*':
        print("Result:", calc.multiply(a, b))

    elif op == '/':
        print("Result:", calc.divide(a, b))

    else:
        raise ValueError("Invalid operator.")

except ValueError as e:
    print("Error:", e)

except ZeroDivisionError as e:
    print("Error:", e)
```

<h2>2) Write a program to check whether a person is eligible to donate blood. Takes age and weight from the user. Blood donation 
criteria are age should be greater than 18 years and the weight should be greater than 50kg. Raise the ValueError exception 
for invalid input and create a custom exception if the above-mentioned criteria don’t match. (A) </h2>

```python
class NotEligibleError(Exception):
    pass

try:
    age = int(input("Enter age: "))
    weight = float(input("Enter weight (kg): "))

    if age <= 0 or weight <= 0:
        raise ValueError("Age and weight must be positive numbers.")

    if age < 18 or weight < 50:
        raise NotEligibleError("Person is NOT eligible to donate blood.")

    print("Person IS eligible to donate blood.")

except ValueError as e:
    print("Invalid Input:", e)

except NotEligibleError as e:
    print("Eligibility Error:", e)
```

<h2>
3) Write a program that divides two numbers and raises a custom exception if the user tries to divide by zero. (B) </h2>

```python
class DivisionByZeroError(Exception):
    pass

try:
    a = float(input("Enter numerator: "))
    b = float(input("Enter denominator: "))

    if b == 0:
        raise DivisionByZeroError("You cannot divide by zero.")

    print("Result:", a / b)

except DivisionByZeroError as e:
    print("Error:", e)
except ValueError:
    print("Invalid number format.")
```
